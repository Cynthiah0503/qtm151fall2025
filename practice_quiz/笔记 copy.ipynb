```python
#linspace
x = np.linspace(0, 2*np.pi, 100)   # from 0 to 2π, 100 points
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#For-loop plotting
import matplotlib.pyplot as plt

list_vectors = [y1, y2]
for vec in list_vectors:
    plt.plot(x, vec)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine and Cosine")
plt.legend(["sin(x)", "cos(x)"])
plt.show()
```

```python
# ============================================================
# 📝 Exam Notes – Detailed Cheatsheet with Key Comments
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ------------------------------------------------------------
# 1. if/elif/else block to test membership in two lists
# ------------------------------------------------------------
fruit = "apple"
list_A = ["apple", "banana"]
list_B = ["orange", "pear"]

# ⚠️ 易错点: 必须用 if/elif/else，不要写多个 if
if fruit in list_A:
    print(fruit, "belongs to List A")
elif fruit in list_B:
    print(fruit, "belongs to List B")
else:
    print(fruit, "does not belong to either list")

# ------------------------------------------------------------
# 2. From two lists -> NumPy array -> histogram
# ------------------------------------------------------------
x_vals = [1, 2, 3, 4]
y_vals = [2, 4, 6, 8]

# ⚠️ 易错点: 必须把 list 转成 np.array 才能逐元素相乘
arr = np.array(x_vals) * np.array(y_vals)

# ⚠️ 易错点: 题目常要求“change the color” -> color="green"
plt.hist(arr, bins=5, color="green")   
plt.xlabel("Value")    # ⚠️ 一定要写 axis labels
plt.ylabel("Frequency")
plt.title("Histogram of x*y")  # ⚠️ Title 必须有
plt.show()

# ------------------------------------------------------------
# 3. Random samples from two distributions -> side-by-side histograms
# ------------------------------------------------------------
# ⚠️ 易错点: 用 np.random.normal / np.random.uniform
sample1 = np.random.normal(loc=0, scale=1, size=1000)   # mean=0, std=1
sample2 = np.random.uniform(low=-2, high=2, size=1000)  # uniform range

# ⚠️ 易错点: side by side 要用 subplots(1,2)
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].hist(sample1, bins=20, color="blue")
axes[0].set_title("Normal(0,1)")
axes[0].set_xlabel("Value")     # 每个子图都要单独 label
axes[0].set_ylabel("Frequency")

axes[1].hist(sample2, bins=20, color="red")
axes[1].set_title("Uniform(-2,2)")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")

plt.tight_layout()   # ⚠️ 避免子图重叠
plt.show()

# ------------------------------------------------------------
# 4. Read CSV -> DataFrame -> scatter plot
# ------------------------------------------------------------
df = pd.read_csv("example.csv")   # ⚠️ 必须写 pd.read_csv
plt.scatter(df["col_x"], df["col_y"], color="purple")  # ⚠️ scatter vs plot
plt.xlabel("col_x")   # ⚠️ 要和 DataFrame 列名一致
plt.ylabel("col_y")
plt.title("Scatter Plot of col_x vs col_y")
plt.show()

# ------------------------------------------------------------
# 5. Loop over list of columns -> print means
# ------------------------------------------------------------
columns = ["col1", "col2", "col3"]
# ⚠️ 易错点: 用 df[col].mean()，不要 np.mean(df[col]) （两者都行，但老师常看这个）
for col in columns:
    print("Mean of", col, "=", df[col].mean())

# ------------------------------------------------------------
# 6. List of CSVs -> loop -> store means in a list
# ------------------------------------------------------------
filenames = ["file1.csv", "file2.csv", "file3.csv"]
means_list = []

# ⚠️ 易错点: for 循环里必须 append，不要直接覆盖
for fname in filenames:
    temp_df = pd.read_csv(fname)
    mean_val = temp_df["target_col"].mean()  # ⚠️ target_col 要换成题目指定的列
    means_list.append(mean_val)

print("Means from all files:", means_list)
```
